```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 1, load the data using pandas

df = pd.read_csv('practical.csv')
df.head()
```

```python
# Step 2, plot the data using seaborn
sns.pairplot(df)
```

```python
# Step 3, have a look at 'A' on its oown
plt.plot(df['A'])
# it looks like there might be an outliers
plt.scatter(0,np.mean(df['A']), marker='x')
```

```python
# filter out values that are too outliers
df = df[df['A']<np.mean(df['A'])]
sns.pairplot(df)
```

```python
# Step 4, look at what is wrong with 'B'
df['B']
# since 'A' was a float, we have a hint that 'B' as an object might be a problem
```

```python
# Here is a trick, you the set function to see what is inside this
set(df['B'])
```

```python
# Step 5, lets filter out by 'NaN and nonnull
df = df[df['B']!="'NaN'"]
df = df[df['B'].notnull()]
```

```python
# note that it is still a string type
set(df['B'])
```

```python
# make the column a float type
df['B'] = df['B'].astype(float)
```

```python
sns.pairplot(df)
```

```python
# Step 6 , filter A by B
sns.pairplot(df[df['B']==0])
```

```python
sns.pairplot(df[df['B']==1])
```

```python
sns.pairplot(df[df['B']==-1])
```
